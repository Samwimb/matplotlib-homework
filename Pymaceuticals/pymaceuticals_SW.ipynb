```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset
clinical_trial_complete = pd.merge(clinical_data, mouse_data, how="left", on=["Mouse ID"])

# Display the data table for preview
clinical_trial_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mtv = clinical_trial_complete.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
mtv_df = pd.DataFrame(mtv)

#resetting the index
mtv_df.reset_index(inplace=True)

# Preview DataFrame
mtv_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoin
stv = clinical_trial_complete.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
stv_df = pd.DataFrame(stv)

#reset Index
stv_df.reset_index(inplace=True)

# Preview DataFrame
stv_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mtv_format_df = mtv_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
stv_format_df = stv_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")


# Preview that Reformatting worked
mtv_format_df.head()
#stv_format_df.head()
```

```python
# Generate the Plot (with Error Bars)

#plot1 = mtv_format_df[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(kind="line", figsize=(15,15), title="Tumor Response to Treatment", grid=True, yerr=stv_format_df, marker="o")
#plot1.set_xlabel("Timepoint")
#plot1.set_ylabel("Mean Tumor Volume")

settings = {
    "Capomulin": {"marker":"o", "color":"red", "linestyle": "--"},
    "Infubinol": {"marker":"^", "color":"green", "linestyle": "--"},
    "Ketapril": {"marker":"s", "color":"blue", "linestyle": "--"},
    "Placebo": {"marker":"D", "color":"black", "linestyle": "--"}
}

fig, ax = plt.subplots()

for key, grp in mtv_df.loc[mtv_df["Drug"].isin(["Capomulin", "Infubinol", "Ketapril", "Placebo"])].groupby(['Drug']):
    ax = grp.plot(ax=ax, figsize=(15,15), kind='line', x="Timepoint", y='Tumor Volume (mm3)', 
                  title="Tumor Response to Treatment", grid=True, yerr=stv_format_df,
                  marker=settings[key]["marker"],
                  c=settings[key]["color"],
                  linestyle=settings[key]["linestyle"],
                  label=key)
    
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='best')
plt.show()

# Save the Figure
plt.savefig("tumor_reponse_time_to_treatment.png")

```

## Metastatic Response to Treatment

```python
# Generate the Plot (with Error Bars)

#plot2 = mms_format_df[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(kind="line", figsize=(15,15), title="Metastatic Spread During Treatment", grid=True, yerr=sms_format_df, marker="o")
#plot2.set_xlabel("Timepoint")
#plot2.set_ylabel("Metastatic Spread")


settings = {
    "Capomulin": {"marker":"o", "color":"red", "linestyle": "--"},
    "Infubinol": {"marker":"^", "color":"green", "linestyle": "--"},
    "Ketapril": {"marker":"s", "color":"blue", "linestyle": "--"},
    "Placebo": {"marker":"D", "color":"black", "linestyle": "--"}
}

fig, ax1 = plt.subplots()

for key, grp in mms_df.loc[mms_df["Drug"].isin(["Capomulin", "Infubinol", "Ketapril", "Placebo"])].groupby(['Drug']):
    ax1 = grp.plot(ax=ax1, figsize=(15,15), kind='line', x="Timepoint", y='Metastatic Sites', 
                  title="Metastatic Spread During Treatment", grid=True, yerr=sms_format_df,
                  marker=settings[key]["marker"],
                  c=settings[key]["color"],
                  linestyle=settings[key]["linestyle"],
                  label=key)
    
plt.xlabel("Time(Days)")
plt.ylabel("Met Sites")
plt.legend(loc='best')
plt.show()

# Save the Figure
plt.savefig("metastatic_spread_during_treatment.png")


```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
per_change = ((mtv_format_transposed_df[45] - mtv_format_transposed_df[0])/mtv_format_transposed_df[0])*100

# Display the data to confirm
per_change

```

```python
# Store all Relevant Percent Changes into a Tuple
x = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
per_change = per_change[["Capomulin", "Infubinol", "Ketapril", "Placebo"]]

# Splice the data between passing and failing drugs

passing = [True if drug <= 0 else False for drug in per_change]
colors = ["g" if drug == True else "r" for drug in passing]
height = [-7 if drug == True else 5 for drug in passing]

# Save the Figure
fig, ax = plt.subplots()

ax.grid(True)
plt.yticks([-20, 0, 20, 40, 60])
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")
ax.bar(x, per_change, width = -1, align="edge", color=colors, edgecolor='black')
for i in range(len(per_change)):
    plt.text(-.5 + i, height[i], str(round(per_change[i]))+ "%", ha="center", color="white", fontweight="bold")
```

Observable Trends:

1) Of the four main drugs (Capomulin, Infubinol, Ketapril, and Placebo) that we displayed in our plots, Capomulin was the only cancer drug that appears to reduce tumor volume in mice.

2) Beyond the four main drugs (Capomulin, Infubinol, Ketapril, and Placebo), Ramicane also appears to reduce tumor volume in mice.

3) Capomulin also was the drug that appeared to be the best at limiting metastatic spread during treatment.
